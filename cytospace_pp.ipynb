## preparing data for cytospace package of deconvolution 

```python
import pandas as pd
import scipy.io
from scipy.sparse import csr_matrix
```

```python
# Paths
matrix_path = "/kaggle/working/data/1160920F_filtered_count_matrix/matrix.mtx"
barcodes_path = "1160920F_filtered_count_matrix/barcodes.tsv"
features_path = "1160920F_filtered_count_matrix/features.tsv"
```

```python
!mv /kaggle/working/data/1160920F_filtered_count_matrix/barcodes.tsv.gz /kaggle/working/data/1160920F_filtered_count_matrix/barcodes.tsv
```

```python
matrix = scipy.io.mmread(matrix_path).tocsc()
```

```python
barcodes = pd.read_csv(barcodes_path, header=None)[0].tolist()
features = pd.read_csv(features_path, sep='\t', header=None)
```

```python
gene_names = features[0].tolist()
```

```python
df = pd.DataFrame.sparse.from_spmatrix(matrix, index=gene_names, columns=barcodes)
```

```python
df = df[~df.index.duplicated(keep='first')]
```

```python
df
```

```python
# Reset index to move gene names to a column
df.reset_index(inplace=True)
df.rename(columns={"index": "Gene"}, inplace=True)
```
